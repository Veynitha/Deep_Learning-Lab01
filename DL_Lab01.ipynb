3. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].

```python
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5,5)
y = np.linspace(-5, 5)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Select the relevant features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]

# Calculate the Pearson correlation matrix
pearson_corr = data.corr(method='pearson')

# Calculate the Spearman correlation matrix
spearman_corr, _ = spearmanr(data)

# Convert Spearman correlation matrix to a DataFrame
spearman_corr_df = pd.DataFrame(spearman_corr, index=features, columns=features)

# Plot Pearson correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation Heatmap')
plt.show()

# Plot Spearman correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr_df, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation Heatmap')
plt.show()
```

3. Calculate the mean, standard deviation and quartile values for each of the above 4 features.


```python
N_total_mean = df1['N_total'].mean()
N_total_std = df1['N_total'].std()
N_total_quantiles = df1['N_total'].quantile([0.25, 0.5, 0.75])

N_bulk_mean = df1['N_bulk'].mean()
N_bulk_std = df1['N_bulk'].std()
N_bulk_quantiles = df1['N_bulk'].quantile([0.25, 0.5, 0.75])

N_surface_mean = df1['N_surface'].mean()
N_surface_std = df1['N_surface'].std()
N_surface_quantiles = df1['N_surface'].quantile([0.25, 0.5, 0.75])

R_avg_mean = df1['R_avg'].mean()
R_avg_std = df1['R_avg'].std()
R_avg_quantiles = df1['R_avg'].quantile([0.25, 0.5, 0.75])

# Output the calculated values
print("N_total mean:", N_total_mean)
print("N_total standard deviation:", N_total_std)
print("N_total quantiles:\n", N_total_quantiles)

print("N_bulk mean:", N_bulk_mean)
print("N_bulk standard deviation:", N_bulk_std)
print("N_bulk quantiles:\n", N_bulk_quantiles)

print("N_surface mean:", N_surface_mean)
print("N_surface standard deviation:", N_surface_std)
print("N_surface quantiles:\n", N_surface_quantiles)

print("R_avg mean:", R_avg_mean)
print("R_avg standard deviation:", R_avg_std)
print("R_avg quantiles:\n", R_avg_quantiles)
```

4. Plot the histogram of each of these features in a 1x4 layout

```python
import pandas as pd
import matplotlib.pyplot as plt


# Step 3: Plot histograms in a 1x4 layout
fig, axs = plt.subplots(1, 4, figsize=(20, 5))

columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
for i, col in enumerate(columns):
    axs[i].hist(df1[col], bins=30, edgecolor='black')
    axs[i].set_title(col)
    axs[i].set_xlabel('Value')
    axs[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

g = sns.PairGrid(df1)


g.map_upper(sns.histplot) #bivariate histogram
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)

# Show the plot
plt.show()
```
